```python
import pandas as pd

markets = {
    "Nasdaq": "data/Nasdaq/NQH20.csv",
    "EuroStoxx": "data/EuroStoxx/VGH22.csv",
    "GBP": "data/GBP - British Pound/BPM20.csv",
    "German Bunds": "data/German Bunds - German Government Bonds/RXM25.csv",
    "Heating Oil": "data/HeatingOil/HOF22.csv",
    "JPY": "data/JPY - Japanese Yen/JYH25.csv"
}

for market, path in markets.items():
    df = pd.read_csv(path, header=None,
                     names=["datetime", "open", "high", "low", "close", "volume"])
    prices = pd.concat([df["open"], df["high"], df["low"], df["close"]])
    prices = prices.sort_values().unique()
    diffs = pd.Series(prices).diff().dropna()
    tick_size = diffs[diffs > 0].min()
    print(f"{market}: {tick_size}")
```
